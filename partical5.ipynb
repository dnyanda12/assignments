```python
ee.Initialize(project='ee-dnyandagawathe24')
```

```python
import geemap.geemap as geemap
```

```python
m = geemap.Map()
m
```

```python
India_District = ee.FeatureCollection('projects/ee-dnyandagawathe24/assets/Indian_district')
m.addLayer(India_District,{},'India_District')
m
```

```python
Visualize_Radiation24 = {"min":217.3,"max":655.7,'palette':['#BF247A','#812B8C','#2A2359','#D9731A','#BF3939']}
```

```python
Radiation24 = ee.ImageCollection("MODIS/061/MCD18A1").filterDate("2024-01-01","2024-12-31").select("Diffuse").mean().clip(India_District)
m.addLayer(Radiation24,Visualize_Radiation24,'Radiation24')
m
```

```python
import os
```

```python
out_dir = os.path.join(os.path.expanduser('~'),'Downloads')
Radiation24_csv = os.path.join(out_dir,'Radiation24.csv')
if not os.path.exists(out_dir):
    os.makedirs(out_dir)
geemap.zonal_stats(Radiation24, India_District, Radiation24_csv, statistics_type='MEAN', scale=300)
```
